```python
cols_null_more_than_0 = null_percent[null_percent > 0.0000].index

# Filter out columns that actually exist in 'df'
existing_columns = [col for col in cols_null_more_than_0 if col in df.columns]

# Now, get the data types for these existing columns
column_datatypes = df[existing_columns].dtypes
print(column_datatypes)
```

all the null value columns are object type thus we can use the mode imputation 

```python
columns_to_impute = ['Mocodes', 'Vict Sex', 'Vict Descent', 'Premis Desc']

for col in columns_to_impute:
    mode_value = df[col].mode().iloc[0]  # Getting the mode for each column
    df[col].fillna(mode_value, inplace=True)  # Imputing missing values with mode
```

### Cross validations

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier

# Initialize the RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=50, random_state=23)

# Perform cross-validation
cv_scores = cross_val_score(rf_classifier, X_train, y_train, cv=5)  # You can adjust the number of folds (cv) as needed

# Print accuracy for each fold
for i, score in enumerate(cv_scores, 1):
    print(f'Fold {i} Accuracy: {score}')

# Train the classifier on the entire training set
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the performance on the test set
Random_forest_accuracy = accuracy_score(y_test, y_pred)
classification_report_str = classification_report(y_test, y_pred)

print(f'\nAccuracy on Test Set: {Random_forest_accuracy}')
```
